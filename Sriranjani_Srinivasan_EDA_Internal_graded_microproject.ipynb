4. Clean the data (Data cleaning) [8 Marks]
a) credit_score column gives a range, Replace the range with its average. Ex= 765-769, 
replace 576 and 796 with = (576+796)/2. [Use str.split and average related functions].


```python
#First we need to find the average
```

```python
def calculate_avg(range_str):
    n= range_str.split('-')
    a= int(''.join(filter(str.isdigit,n[0])))
    b= int(''.join(filter(str.isdigit,n[1])))
    average = (a+b)/2
    return average
df['credit_score'] = df['credit_score'].apply(calculate_avg)
df['credit_score']
```

```python
df
```

```python
df.info()
```

```python
df['Interest.Rate(in %)'] = df['Interest.Rate(in %)'].str.replace('#','')
df['Interest.Rate(in %)'] = pd.to_numeric(df['Interest.Rate(in %)'])
```

```python
df['Loan duration (months)'] = df['Loan duration (months)'].replace(['>= 36 months'],'36 months')
df['Loan duration (months)'] = df['Loan duration (months)'].str.replace('months','')
df['Loan duration (months)'] = df['Loan duration (months)'].replace(['.','NA',''],0)
df['Loan duration (months)'] = pd.to_numeric(df['Loan duration (months)'])


```

```python
df
```

```python
df['Debt.To.Income.Ratio'] = df['Debt.To.Income.Ratio'].str.replace('!','')
df['Debt.To.Income.Ratio'] = df['Debt.To.Income.Ratio'].replace(['NA',''],0)
df['Debt.To.Income.Ratio'] = df['Debt.To.Income.Ratio'].astype(float)
```

```python
df['Family_incomeIncome'] = pd.to_numeric(df['Family_incomeIncome'].replace(['>=6250', '<8333.33>', '< 7083.33', 
                                                                             '4416.67 !', '< 8833.33>', '<4000 >', 
                                                                             'NA', ''], ['6250', '8333.33', '7083.33', 
                                                                                    '4416.67', '8833.33', '4000', 0, 0]))
df['Family_incomeIncome']
```

5. Check for outliers present in the data, perform outlier analysis using the IQR method if any 
outliers are present.

```python
plt.figure(figsize = (30,10))
sns.boxplot(data=df)
```

```python
#Outlier removal using IQR amt requested for loan
percentile25 = df['Amount.Requested for loan'].quantile(0.25)
percentile75 = df['Amount.Requested for loan'].quantile(0.75)

IQR = percentile75 - percentile25
IQR
```

```python
Upper_Limit = percentile75 + 1.5*IQR
Upper_Limit
```

```python
Lower_Limit = percentile25 - 1.5*IQR
Lower_Limit
```

```python
df = df[df['Amount.Requested for loan']  < Upper_Limit]
df
```
